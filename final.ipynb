```python
import matplotlib.pyplot as plt
import numpy as np
import platform  # pyplot 한글 깨짐 문제

label = ['20대', '30대', '40대', '50대', '60대']
res = [94.66, 89.89, 87.82, 86.77, 82.11]

# Basic Bar Chart
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 8))
plt.bar(label, res, width=0.8)
plt.xticks(fontsize=15)
plt.title("외식비 이용 비율", fontsize=25)
plt.xlabel('나이대별', fontsize=18)
plt.ylabel('이용 비율', fontsize=18)
plt.show()
```
