```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import platform  # pyplot 한글 깨짐 문제
```

```python
df = pd.read_csv('./temp2.csv')
```

```python
# 조건 행 추출, num 나이 앞 자리수
def col_by_age(num):
    temp_df = df[(df['Age'] == num * 10) | (df['Age'] == num * 10 + 9)]

    return temp_df


def get_dropout(ratio_np, total_np):
    total_sort_des = np.argsort(total_np)[::-1]
    print("20대 최대 총가계지출: ", total_np[total_sort_des][:10])
    print("20대 최대 총가계지출 index: ", total_sort_des[:10])
    dropout = int(len(ratio_np) * 0.005)

    food_ratio_list = list(ratio_np[total_sort_des][dropout:-1])
    total_list = list(total_np[total_sort_des][dropout:-1])

    return food_ratio_list, total_list


def get_ratio(df, colName):
    _20_df_ratio = []
    for i in df.index.tolist():
        _20_df_ratio.append(df.loc[i, colName] / df.loc[i, "총가계지출"] * 100)

    return _20_df_ratio


def draw_scatter(x_axis, y_axis, title, y_title):
    # x축 식료품비 비율
    # y축 총가계지출
    plt.rc('font', family='Malgun Gothic')
    plt.figure(figsize=(10, 10))
    plt.scatter(x_axis, y_axis, s=3, alpha=0.5)
    plt.title(title, fontsize=20)
    plt.xlabel('총가계지출', fontsize=15)
    plt.ylabel(y_title, fontsize=15)
    plt.show()
```

```python
_20_df = col_by_age(2)
_20_df_ratio = get_ratio(_20_df, "교통/통신비")
# _20_df_ratio = _20_df['교통/통신비']

_20_ratio_list, _20_total_list = get_dropout(
    np.array(_20_df_ratio), np.array(_20_df["총가계지출"]))
draw_scatter(_20_total_list, _20_ratio_list, "20대", "교통/통신비")
```
